```python
from PIL import Image
import os

def resize_images(input_folder, output_folder, new_size):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop through all files in the input folder
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        # Check if the file is an image
        if os.path.isfile(input_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
            # Open the image
            img = Image.open(input_path)

            # Resize the image
            img_resized = img.resize(new_size)

            # Save the resized image to the output folder
            img_resized.save(output_path)

            print(f"Resized: {filename}")

if __name__ == "__main__":
    # Set your input and output folders
    input_folder = "J:\weld pool images_bib\V6"
    output_folder = "J:\weld pool images_bib\resizedV6"

    # Set the new size (width, height)
    new_size = (500, 500)

    # Resize images
    resize_images(input_folder, output_folder, new_size)
```

```python
import cv2
import os

input_dir = r"J:\weld pool images_bib\V6"
output_dir = r"J:\weld pool images_bib\resizedV6"

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image
        img = cv2.imread(os.path.join(input_dir, filename))

        # Check if the image is valid
        if img is not None and img.size > 0:
            # Set the desired width and height
            desired_width = 500
            desired_height = 500

            # Resize the image
            resized_img = cv2.resize(img, (desired_width, desired_height))

            # Save the resized image
            cv2.imwrite(os.path.join(output_dir, filename), resized_img)
        else:
            print(f"Invalid image: {filename}")
```
