```python
import os
import shutil
import random
from tqdm import tqdm
```

```python
Dataset_root = r'J:\position detection\640-480 normal welding image'
```

```python
os.chdir(os.path.join(Dataset_root, 'images'))
```

```python
os.listdir()
```

```python
test_frac = 0.2  # rate of validation
random.seed(123) # random seed
```

```python
folder = '.'
```

```python
img_paths = os.listdir(folder)
random.shuffle(img_paths) # random distribute

val_number = int(len(img_paths) * test_frac) # volume of validation
train_files = img_paths[val_number:]         
val_files = img_paths[:val_number]           

print('totalnumber of datasets', len(img_paths))
print('training volume', len(train_files))
print('validation volumn', len(val_files))
```

Moving the training images to train folder

```python
os.mkdir('train')
for each in tqdm(train_files):
    shutil.move(each, 'train')
```

Moving the validation images to validation folder

```python
os.mkdir('val')
for each in tqdm(val_files):
    shutil.move(each, 'val')
```

```python
os.chdir('../json')
```

```python
os.mkdir('train')
for each in tqdm(train_files):
    srt_path = each.split('.')[0] + '.json'
    shutil.move(srt_path, 'train')
```

```python
os.mkdir('val')
for each in tqdm(val_files):
    srt_path = each.split('.')[0] + '.json'
    shutil.move(srt_path, 'val')
```
